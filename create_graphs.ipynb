```python
import json

import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import pyplot as plt

from pprint import pprint

import plotly.io as pio
pio.renderers.default='jupyterlab'  # notebook doesn't work check https://plotly.com/python/renderers/
pd.options.display.max_columns = 100
pd.options.display.max_rows = 300
```

```python
plt.style.use(['science',"vibrant"])
```

```python
plot_levels = pd.read_csv("plot_levels.csv")
```

```python
plot_levels = plot_levels.rename(columns={"CATEGORY_LEVEL":"CATEGORY LEVEL"})
```

```python
sns.barplot(data=plot_levels, x="CATEGORY LEVEL",y="count",dodge=False)
plt.savefig("levels_freq.png",dpi=600)
```

```python
plot_largest = pd.read_csv("plot_largest.csv")
```

```python
plot_largest = plot_largest.rename(columns={"CATEGORY_LEVEL":"CATEGORY LEVEL",
                                         "CATEGORY_PATH":"CATEGORY PATH"})
```

```python
plot_largest["CATEGORY PATH"] = plot_largest["CATEGORY PATH"].apply(lambda x: x.replace(" > "," $>$ "))
```

```python
sns.barplot(data=plot_largest, hue="CATEGORY LEVEL",
            y="CATEGORY PATH",x="count", dodge=False)
plt.xticks(rotation=90);
plt.ylabel("")
plt.legend(frameon=True,title="CATEGORY LEVEL")
plt.savefig("largest_categories.png",dpi=600)
```

```python
plot_mat_counts = pd.read_csv("plot_mat_counts.csv")
```

```python
plot_mat_counts = plot_mat_counts.rename(columns={"MATERIAL_ID":"MATERIAL ID",})
```

```python
plot_mat_counts = plot_mat_counts["MATERIAL ID"].value_counts().reset_index()
```

```python
sns.barplot(data=plot_mat_counts,y="MATERIAL ID",x="index")
plt.ylabel("count")
plt.xlabel("number of occurences in dataset")
plt.savefig("plot_mat_counts.png",dpi=600)
```

```python
plot_counts = pd.read_csv("plot_counts.csv")
```

```python
plot_counts = plot_counts.rename(columns={"CATEGORY_ID":"CATEGORY ID",
                                         "CATEGORY_PATH":"CATEGORY PATH"})
```

```python
sns.histplot(data=plot_counts,x="CATEGORY ID", bins=100)
plt.xlabel("number of materials in category");
plt.ylabel("count of categories");
plt.savefig("plot_counts.png",dpi=600)
```

```python
sns.histplot(data=plot_counts[plot_counts["CATEGORY ID"] <1000],
x="CATEGORY ID", bins=100)
plt.xlabel("number of materials in category");
plt.ylabel("count of categories");
plt.savefig("plot_counts_detail.png",dpi=600)
```

```python
history = pd.read_csv("CNN_128.csv")
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1, 2, 1)
plt.plot(history['accuracy'])
plt.plot(history['val_accuracy'])
# plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
# plt.legend(['train', 'val'], loc='lower left', bbox_to_anchor=(1, 0.75), frameon=True)
# plt.savefig("accuracy_plot.png",dpi=600)
# plt.show()

plt.subplot(1, 2, 2)
plt.plot(history['loss'])
plt.plot(history['val_loss'])
# plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right', frameon=True, )
plt.savefig("overfit128.png",dpi=600)
```
